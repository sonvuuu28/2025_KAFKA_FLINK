# Test KAFKA

## I. Common command

### 1. List các topic

```bash
kafka-topics --bootstrap-server localhost:9092 --list
```

### 2. Tạo topic

```bash
docker exec -it broker bash
```

```bash
kafka-topics --create \
  --topic test \
  --bootstrap-server broker:29092 \
  --partitions 1 \
  --replication-factor 1
```

### 3. Xóa topic

```bash
docker exec -it broker bash
```

```bash
kafka-topics --bootstrap-server localhost:9092 --delete --topic test
```
